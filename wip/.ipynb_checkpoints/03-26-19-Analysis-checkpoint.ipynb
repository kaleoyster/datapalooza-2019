```python
df2['Percentage Male'] = df2['Female; Estimate; Total population'] / df2['Total; Estimate; Total population']
df['Percentage Uninsured'] = df['Total; Estimate; Total civilian noninstitutionalized population'] / df['Total Uninsured; Estimate; Total civilian noninstitutionalized population']
```

```python
 df2['total male population'] = df2['Total; Estimate; Total population'] * df2['Percent female; Estimate; Total population']
```

```python
df2['Female; Estimate; Total population'] / df2['Total; Estimate; Total population']
```

```python
d_male = {}
for county, gender_count in zip(df2["Geography"],df2["Percentage Male"]):
    d_male[county] = gender_count
    #print(county, gender_)
```

```python
d = {}
for county, total_unin in zip(df["Geography"], df["Percentage Uninsured"]):
    d[county] = total_unin
    #print(county, total_unin)
```

```python
gender = []
unin = []
for county in d.keys():
    gender.append(d_male[county])
    unin.append(d[county])
```

```python
import researchpy as rp
gender_cat = pd.qcut(gender,[0, .25, .75, 1.], labels = ['High','Medium','Low'])
unin_cat = pd.qcut(unin,[0, .25, .75, 1.], labels = ['High','Medium','Low'])

d = pd.DataFrame({"Male":gender_cat, "Uninsured":unin_cat})
#d = d[d['Male'].isin(['High','Low'])]
#d = d[d['Uninsured'].isin(['High','Low'])]

table, results = rp.crosstab(d['Male'], d['Uninsured'], prop= 'col', test= 'chi-square')
    
#table
```

```python
y = df['percent insured']
X = df[df_estimate_column[2:56]]
## fit a OLS model with intercept on TV and Radio
X = sm.add_constant(X)
est = sm.OLS(y.astype(float),X.astype(float)).fit()
est.summary()
```

```python
xt = pd.DataFrame(est.params)
xt.columns = [['Coeffs']]
```

```python
xt['abs Coeffs'] = xt['Coeffs'].abs()
```
